鐵條切割問題
擷取.PNG

```
def BottomUpCutRod(p, n):
    r = [0]*(n+1)
    for i in range(1, n+1):
        if n == 0:
            return 0
        q =0
        for j in range(1, i+1):
            q = max(q, p[j]+r[i-j])
            r[i] = q
    return r[n]
p=[0,1,5,8,9,10,17,17,20,24,30]
asd=int(input("請輸入0~10的整數"))
print("長度為",asd,"的鐵條切割後可賣出最高總價為",BottomUpCutRod(p,asd))  
```
