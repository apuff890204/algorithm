Selection sort (選擇排序)

```
def Selction_Sort(datas):
    for i in range(len(datas)-1):           # [(n-1)+1]
        min = i                             # 第一次選從第一個數開始,第二次從第二個數開始
        for j in range(i+1, len(datas)):    # 用記錄的max數與其他的比較
            if datas[j] < datas[min]:
                min = j

        datas[i], datas[min] = datas[min], datas[i]     #(n-1)

        print(datas)                                    #(n-1)
        print('---'*16)                                 #(n-1)    
    return datas                                        # 1
                                                        # O(n^2)

if __name__ == '__main__':
    import random
    a = list(range(16))
    random.shuffle(a)
    print(a)
    a = Selction_Sort(a)
    print(a)
```
